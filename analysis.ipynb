```python
import sqlite3, pandas as pd

conn = sqlite3.connect("ecommerce_orders.db")

# sanity check: list tables
pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
```

```python
# Top 10 countries by revenue
q1 = """
SELECT Country,
       ROUND(SUM(TotalValue),2) AS Revenue,
       COUNT(DISTINCT InvoiceNo) AS Orders
FROM orders
GROUP BY Country
ORDER BY Revenue DESC
LIMIT 10;
"""
top_countries = pd.read_sql_query(q1, conn)
top_countries
```

```python
q2 = """
WITH country_orders AS (
  SELECT Country, COUNT(DISTINCT InvoiceNo) AS Orders
  FROM orders GROUP BY Country
),
country_returns AS (
  SELECT Country, COUNT(DISTINCT InvoiceNo) AS Returns
  FROM returns GROUP BY Country
)
SELECT o.Country,
       o.Orders,
       COALESCE(r.Returns,0) AS Returns,
       ROUND(1.0*COALESCE(r.Returns,0)/o.Orders,4) AS ReturnRate
FROM country_orders o
LEFT JOIN country_returns r USING (Country)
WHERE o.Orders >= 100
ORDER BY ReturnRate DESC
LIMIT 10;
"""
return_rate = pd.read_sql_query(q2, conn)
return_rate
```

```python
q3 = """
SELECT InvoiceHour, COUNT(DISTINCT InvoiceNo) AS Orders
FROM orders
GROUP BY InvoiceHour
ORDER BY InvoiceHour;
"""
by_hour = pd.read_sql_query(q3, conn)
by_hour
```

```python
top_countries.to_csv("top_countries.csv", index=False)
return_rate.to_csv("return_rate_by_country.csv", index=False)
by_hour.to_csv("orders_by_hour.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Revenue by country (Top 10)
top = pd.read_csv("top_countries.csv")
plt.figure(figsize=(8,4))
plt.bar(top["Country"], top["Revenue"])
plt.xticks(rotation=45, ha="right")
plt.title("Top Countries by Revenue")
plt.ylabel("Revenue")
plt.tight_layout()
plt.savefig("viz_top_countries_revenue.png", dpi=200)
plt.show()

# 2) Return rate by country (Top 10 in your table)
ret = pd.read_csv("return_rate_by_country.csv")
plt.figure(figsize=(8,4))
plt.bar(ret["Country"], ret["ReturnRate"])
plt.xticks(rotation=45, ha="right")
plt.title("Return Rate by Country")
plt.ylabel("Return Rate")
plt.tight_layout()
plt.savefig("viz_return_rate_country.png", dpi=200)
plt.show()

# 3) Orders by hour
byh = pd.read_csv("orders_by_hour.csv")
plt.figure(figsize=(8,4))
plt.plot(byh["InvoiceHour"], byh["Orders"], marker="o")
plt.title("Orders by Hour of Day")
plt.xlabel("Hour"); plt.ylabel("Orders")
plt.tight_layout()
plt.savefig("viz_orders_by_hour.png", dpi=200)
plt.show()
```
